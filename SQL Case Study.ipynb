```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
with sqlite3.connect('sqlite_db_pythonsqlite.db') as con:
    cur = con.cursor()
    cur.execute('select f.name, ((memBookings.slots * f.membercost) + (guestBookings.slots * f.guestcost)) as revenue from Facilities as f inner join (select facid, sum(slots) as slots from Bookings where memid != 0 group by facid) as memBookings using(facid) inner join (select facid, sum(slots) as slots from Bookings where memid = 0 group by facid) as guestBookings using(facid) group by facid order by revenue desc')
    q10 = pd.DataFrame(cur.fetchall())
    q10.columns = [t[0] for t in cur.description]
q10
```

```python
with sqlite3.connect('sqlite_db_pythonsqlite.db') as con:
    cur = con.cursor()
    cur.execute('select m1.surname, m1.firstname, (m2.surname || " " || m2.firstname) as "Recommended By" from Members as m1 inner join Members as m2 on m1.recommendedby = m2.memid order by m1.surname')
    q11 = pd.DataFrame(cur.fetchall())
    q11.columns = [t[0] for t in cur.description]
q11
# FILTERS OUT RECOMMENDED_BY=null rows for some reason
# In PHPMyAdmin it includes those rows and indicates Recommended By=GUEST GUEST, for some reason
```

```python
with sqlite3.connect('sqlite_db_pythonsqlite.db') as con:
    cur = con.cursor()
    cur.execute('select f.name, sum(slots) as "Member Visits" from Bookings inner join Facilities as f using(facid) where memid != 0 group by facid')
    q12 = pd.DataFrame(cur.fetchall())
    q12.columns = [t[0] for t in cur.description]
q12
# Question 12 A- Finds the count of total Member visits per Facility
```

```python
with sqlite3.connect('sqlite_db_pythonsqlite.db') as con:
    cur = con.cursor()
    cur.execute('select strftime("%m", starttime) as Month, sum(slots) as "Facility Visits" from Bookings where memid != 0 group by Month')
    q13 = pd.DataFrame(cur.fetchall())
    q13.columns = [t[0] for t in cur.description]
q13
# Question 13- Returns facilities visits per month- works in PHPMyAdmin w/ EXTRACT, in sqlite, using strftime instead
```
